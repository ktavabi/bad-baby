<h3>Group level sensor domain analysis</h3>

Imports

```python
import os
from os import path as op
import matplotlib.pyplot as plt
import numpy as np
import mne
from mne import (read_evokeds, grand_average)
mne.set_log_level(verbose='WARNING')
%matplotlib inline
```

```python
from mne.stats import permutation_t_test
from mne.viz import plot_evoked_topo


def permute_sensors(epochs, tmin, tmax, channels='grad', nperm=1024, n_jobs=1, alpha=.05):
    """Return sensor array stats as evoked object"""
    # pick MEG Gradiometers
    epochs.pick_types(meg=channels, eeg=False, stim=False, eog=False,
                           include=[], exclude='bads')
    picks = mne.pick_types(epochs.info, meg=channels, eeg=False, stim=False, eog=True,
                           include=[], exclude='bads')
    epochs.drop_bad_epochs()
    # pull out data and times
    data = epochs.get_data()
    times = epochs.times
    # Apply temporal mask to select time window
    temporal_mask = np.logical_and(tmin <= times, times <= tmax)
    data = np.mean(data[:, :, temporal_mask], axis=2)
    # T-test
    T0, p_values, H0 = permutation_t_test(data, n_permutations=nperm, n_jobs=n_jobs)
    # find significant sensors
    significant_sensors = picks[p_values <= alpha]
    print("Number of significant sensors : %d" % len(significant_sensors))
    print("Sensors names : MEG%s" % significant_sensors)
    evoked = mne.EvokedArray(-np.log10(p_values)[:, np.newaxis],
                             epochs.info, tmin=0.)
    return picks, p_values, significant_sensors, evoked
```

### Grand averaged ERF maps - Deviant stimuli
Using `mne.grand_average` to calculate group avraged ERF

#### Group - 2 monthers

```python
evokeds = []
for subj in two_mos:
    evoked_file = op.join(studydir, subj, 'inverse',
                          '%s_%d-sss_eq_%s-ave.fif' % (analysis, lpf,
                                                       subj))
    evoked = read_evokeds(evoked_file, condition=conditions[0], baseline=(None,0))
    evokeds.append(evoked)

two_mos_grndavr = grand_average(evokeds)
```

```python
print('\nN=%d' % len(two_mos))
hs = two_mos_grndavr.plot_joint(ts_args={'gfp' : True}, topomap_args={'outlines' : 'head'})
for h, ch_type in zip(hs, ['grad', 'mag']):
    h.savefig(op.join(studydir, fig_dir, 'twomos-%s_%d_%s_N%d-ave.png'
                      % (conditions[0], lpf, ch_type, len(two_mos))), dpi=240, format='png')
```

#### Group - 6 monthers

```python
evokeds = []
for subj in six_mos:
    evoked_file = op.join(studydir, subj, 'inverse',
                          '%s_%d-sss_eq_%s-ave.fif' % (analysis, lpf,
                                                       subj))
    evoked = read_evokeds(evoked_file, condition=conditions[0], baseline=(None,0))
    evokeds.append(evoked)

six_mos_grndavr = grand_average(evokeds)
```

```python
print('\nN=%d' % len(six_mos))
hs = six_mos_grndavr.plot_joint(ts_args={'gfp' : True}, topomap_args={'outlines' : 'head'})
for h, ch_type in zip(hs, ['grad', 'mag']):
    h.savefig(op.join(studydir, fig_dir, 'sixmos-%s_%d_%s_N%d-ave.png'
                      % (conditions[0], lpf, ch_type, len(six_mos))), dpi=240, format='png')
```

### Grand averaged mismatch responses
Load oddball ERF data, calculate group averages and save to disk as evoked files.

```python
analysis = 'Deviants'
for grp, name in zip([two_mos, six_mos], ['two_mos', 'six_mos']):
    evokeds = []
    for subj in grp:
        evoked_file = op.join(studydir, subj, 'inverse',
                              '%s_%d-sss_eq_%s-ave.fif' % (analysis, lpf,
                                                           subj))
        evoked = read_evokeds(evoked_file, condition='deviants', baseline=(None,0))
        evokeds.append(evoked)
    grndavr = grand_average(evokeds)
    mne.Evoked.save(grndavr, op.join(studydir, '%s_deviants_%d-N%d-ave.fif' 
                                     % (name, lpf, len(grp))))
```

```python
analysis = 'Oddball'
conditions = ['standard', 'Ba', 'Wa']
for cond in conditions:
    for grp, name in zip([two_mos, six_mos], ['two_mos', 'six_mos']):
        evokeds = []
        for subj in grp:
            evoked_file = op.join(studydir, subj, 'inverse',
                                  '%s_%d-sss_eq_%s-ave.fif' % (analysis, lpf,
                                                               subj))
            evoked = read_evokeds(evoked_file, condition=cond, baseline=(None,0))
            assert evoked.comment == cond
            evokeds.append(evoked)
        grndavr = grand_average(evokeds)
        mne.Evoked.save(grndavr, op.join(studydir, '%s_%s_%d-N%d-ave.fif' 
                                         % (name, cond, lpf, len(grp))))
```
